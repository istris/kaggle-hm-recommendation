```python
# Day 1: 数据探索
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置显示选项
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# 加载数据（Kaggle Notebook路径）
DATA_PATH = '/kaggle/input/h-and-m-personalized-fashion-recommendations/'

print("=" * 60)
print("Day 1: H&M数据探索")
print("=" * 60)

# 1. 加载商品数据
print("\n1. 商品数据 (articles.csv)")
articles = pd.read_csv(DATA_PATH + 'articles.csv')
print(f"形状: {articles.shape}")
print(f"字段: {articles.columns.tolist()}")
print("\n前3行:")
display(articles.head(3))

# 关键统计
print(f"\n商品类别分布:")
print(articles['index_group_name'].value_counts())

# 2. 加载用户数据
print("\n" + "=" * 60)
print("2. 用户数据 (customers.csv)")
customers = pd.read_csv(DATA_PATH + 'customers.csv')
print(f"形状: {customers.shape}")
print(f"字段: {customers.columns.tolist()}")
print("\n用户年龄分布:")
print(customers['age'].describe())

# 3. 加载交易数据（只读前100万行，太大）
print("\n" + "=" * 60)
print("3. 交易数据 (transactions_train.csv) - 采样查看")
transactions_sample = pd.read_csv(
    DATA_PATH + 'transactions_train.csv',
    nrows=1000000  # 先读100万行
)
print(f"采样形状: {transactions_sample.shape}")
print(f"字段: {transactions_sample.columns.tolist()}")
print(f"\n时间范围: {transactions_sample['t_dat'].min()} 至 {transactions_sample['t_dat'].max()}")

# 4. 关键洞察：冷启动问题
print("\n" + "=" * 60)
print("4. 冷启动问题分析")

# 计算每个用户的交易次数
user_counts = transactions_sample['customer_id'].value_counts()
print(f"\n用户交易次数分布:")
print(f"  平均: {user_counts.mean():.1f}次")
print(f"  中位数: {user_counts.median():.1f}次")
print(f"  仅购买1次的用户: {(user_counts == 1).sum()} ({(user_counts == 1).mean()*100:.1f}%)")

# 新商品问题
article_counts = transactions_sample['article_id'].value_counts()
print(f"\n商品被购买次数:")
print(f"  仅被购买1次的商品: {(article_counts == 1).sum()} ({(article_counts == 1).mean()*100:.1f}%)")

# 5. 可视化
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 用户年龄分布
axes[0, 0].hist(customers['age'].dropna(), bins=30, edgecolor='black')
axes[0, 0].set_title('User Age Distribution')
axes[0, 0].set_xlabel('Age')

# 用户交易次数分布（log scale）
axes[0, 1].hist(user_counts.values, bins=50, edgecolor='black')
axes[0, 1].set_yscale('log')
axes[0, 1].set_title('User Transaction Count (Log Scale)')
axes[0, 1].set_xlabel('Number of Transactions')

# 商品大类分布
articles['index_group_name'].value_counts().plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Product Category Distribution')
axes[1, 0].tick_params(axis='x', rotation=45)

# 价格分布
axes[1, 1].hist(transactions_sample['price'], bins=50, edgecolor='black')
axes[1, 1].set_title('Price Distribution')
axes[1, 1].set_xlabel('Price')

plt.tight_layout()
plt.savefig('day1_eda.png', dpi=150, bbox_inches='tight')
print(f"\n图表已保存: day1_eda.png")

print("\n" + "=" * 60)
print("Day 1 完成！关键发现：")
print(f"1. 数据集规模: {len(customers)}用户, {len(articles)}商品")
print(f"2. 冷启动问题严重: {(user_counts == 1).mean()*100:.1f}%用户只有1次购买")
print(f"3. 数据时间跨度: {transactions_sample['t_dat'].min()} 至 {transactions_sample['t_dat'].max()}")
print("=" * 60)
```
